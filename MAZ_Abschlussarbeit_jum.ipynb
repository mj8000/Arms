```python
import sipri
import io
from io import StringIO
```

```python
data = sipri.sipri_data(low_year='2014',high_year='2021',seller='',buyer='',armanent_category='any',buyers_or_sellers='',filetype='csv',include_open_deals='on',sum_deliveries='on')
df = pd.read_csv(StringIO(data),keep_default_na=False,na_values=['None'])
df.head(50)
```

```python
#Filter nach allen GeschÃ¤ften, bei denen Saudiarabien das KÃ¤uferland war
```

```python
sa_df = df[df["buyer"] == 'Saudi Arabia'] #HAUPTDATENSET
sa_df.head(50)
sa_df.to_csv('Saudi_weaponstransfers.csv') #Datenset speichern
```

```python
sa_df.head(500)
del sa_df["tidn"]
```

```python
#Weapons categories allgemein verstÃ¤ndlich ausdeutschen gemÃ¤ss der Sipri-Codifizierung: https://www.sipri.org/databases/armstransfers/background
replace_values = {'AC' : "Kampfflugzeuge und Drohnen", 'AD' : "Flugabwehrsysteme", 'AR' : "Artillerie", 'AV' : "gepanzerte Fahrzeuge", 'EN' : "Ersatzteile", 'GR' : "Radarsysteme", 'MI' : "Raketen", 'NW' : "Schiffsbewaffnung", 'OT' : "andere", 'SH' : "Schiffe"}
sa_df = sa_df.replace({"wcat": replace_values})   
```

```python
df_3 = sa_df.groupby(['wcat'])['wcat'].count()
print(df_3)
df_3.plot(kind="bar", xlabel='', ylabel='Anzahl Waffendeals', title='Diese Waffentypen wurden geliefert')
```

```python
df_3 = sa_df.groupby(['seller'])['seller'].count()
print(df_3)
```

```python
sui_df = sa_df[sa_df["seller"] == 'Switzerland']
sui_df.head(10)
sui_df = sui_df.groupby(['wcat'])['wcat'].count()
sui_df.plot(kind="bar",xlabel='', ylabel='Anzahl Waffendeals', title='Die Schweiz macht mit wenigen Deals gutes Geld')
```

```python
fra_df = sa_df[sa_df["seller"] == 'France']
fra_df.head(10)
fra_df = fra_df.groupby(['wcat'])['wcat'].count()
```

```python
fra_df.plot(kind="bar",xlabel='', ylabel='Anzahl Waffendeals', title='Frankreich versorgt Saudiarabien umfassend')
```
